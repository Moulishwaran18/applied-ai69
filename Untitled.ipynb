```python
!pip install ultralytics

import torch
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os

print("PyTorch Version:", torch.__version__)
print("CUDA Available:", torch.cuda.is_available())

model = YOLO("yolov8n.pt")
print("\nYOLOv8 model loaded successfully!")

try:
    from google.colab import files
    uploaded = files.upload()
    img_file = list(uploaded.keys())[0]
    print(f"\nUsing uploaded image: {img_file}")
except:
    print("\nUpload not supported. Using default sample image.")
    img_file = "https://ultralytics.com/images/bus.jpg"

results = model(img_file, conf=0.5)

res = results[0]
bgr = res.plot()
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 8))
plt.imshow(rgb)
plt.title("YOLOv8 Detection Output")
plt.axis("off")
plt.show()

print("\n--- Detection Summary ---")
print(f"Inference Time: {res.speed['inference']:.2f} ms")
print(f"Objects Detected: {len(res.boxes)}\n")

for i, box in enumerate(res.boxes, start=1):
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    label = model.names[cls_id]

    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

    print(f"Object {i}: {label.upper()}")
    print(f"  Confidence : {conf*100:.2f}%")
    print(f"  Bounding Box: ({x1}, {y1}) → ({x2}, {y2})\n")
```
