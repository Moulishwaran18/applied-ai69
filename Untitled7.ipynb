```python
from google.colab import files
import cv2
import numpy as np
import time
from google.colab.patches import cv2_imshow

# Upload image
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

SCALE_FACTOR = 0.026

img = cv2.imread(img_path)
output_img = img.copy()

start = time.time()

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
smooth_img = cv2.GaussianBlur(gray_img, (5, 5), 0)
edge_map = cv2.Canny(smooth_img, 50, 150)

kernel_mat = np.ones((3, 3), np.uint8)
expanded_edges = cv2.dilate(edge_map, kernel_mat, iterations=1)

contour_list, _ = cv2.findContours(expanded_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

object_count = 0

for contour in contour_list:
    region_area = cv2.contourArea(contour)

    if region_area > 500:
        object_count += 1

        x_pos, y_pos, width, height = cv2.boundingRect(contour)

        width_real = round(width * SCALE_FACTOR, 2)

        cv2.rectangle(output_img, (x_pos, y_pos), (x_pos + width, y_pos + height), (0, 255, 0), 2)

        cv2.putText(output_img, f"{width_real} cm",
                    (x_pos, y_pos - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.5,
                    (255, 0, 0),
                    1)

cv2.putText(output_img, f"Count: {object_count}",
            (20, 40),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 0, 255),
            2)

end = time.time()

print("Processing Time:", round(end - start, 3), "sec")

cv2_imshow(output_img)
```
